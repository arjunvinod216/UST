```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize = (15,9))
plt.style.use('seaborn-whitegrid')
plt.title('Letter frequencies')
plt.xlabel('Letters')
plt.ylabel('Frequency')
def wordcount(data):
    d = dict()
    for i in range (len(data)):
        if (data[i].isalpha()):
            if data[i] in d:
                d[data[i]] = d[data[i]] + 1
            else:
                d[data[i]] = 1
    return (d)
  
file = open("constitution.txt", "r")
sen = file.read()
print(wordcount(sen))
pd.Series(wordcount(sen)).plot.bar()
```
